# **Hypothesis Testing**

## **Data Preparation for Variables 'Education' and 'Income'**

```python
df_stat = df1[['Education', 'Income']]
df_stat
```

### **Clean NaN**

NaN Value will be replaced by the mean value of the "Income" variable.

```python
Mean_Income = df_stat["Income"].mean()
print(Mean_Income)
```

```python
df_stat = df_stat.fillna(Mean_Income)
df_stat
```

### **Convert data education into integer code**

```python
np.unique(df_stat["Education"])
```

```python
edu_cat = df_stat['Education'].map({'Basic':1, 'Graduation':2, '2n Cycle': 3, 'Master': 3, 'PhD': 4})
edu_cat = edu_cat.to_frame(name = 'Education_Convert')
edu_cat
```

```python
df_stat_rev = pd.concat([edu_cat, df_stat], axis = 1)
```

```python
df_stat_rev = df_stat_rev.drop(df_stat_rev.columns[[1]], axis=1)
```

```python
df_stat_rev = df_stat_rev.sort_values(by=['Education_Convert'])
df_stat_rev
```

## **Testing**

### **One-Way Anova**

```python
df_Basic = df_stat_rev[df_stat_rev['Education_Convert'] == 1]
df_Graduation = df_stat_rev[df_stat_rev['Education_Convert'] == 2]
df_2nCycle = df_stat_rev[df_stat_rev['Education_Convert'] == 3]
df_Master = df_stat_rev[df_stat_rev['Education_Convert'] == 3]
df_PhD = df_stat_rev[df_stat_rev['Education_Convert'] == 4]
```

```python
anova_test = st.f_oneway(df_2nCycle['Income'], df_Basic['Income'], df_Graduation['Income'], df_Master['Income'], df_PhD['Income'])
pvalue = anova_test.pvalue
pvalue
```

```python
Significance_level = 0.05
```

```python
if pvalue < Significance_level:
  print("'Education' variable is significantly affecting customers' 'Income' variable")
else:
    print("'Education' variable is NOT significantly affecting customers 'Income' variable")
```

### **Covariance**

```python
df_cov = df_stat_rev.cov()
df_cov
```

```python
covar = (df_cov > 0).any()
covar
```

```python
if covar.any() == True:
  print("Due to covariance value being positive (covar > 0), 'Education' variable is positively correlated with 'Income' variable. Means increase in one variable leads to increase on the other one")
else:
  print("Due to covariance value being negative (covar < 0), 'Education' variable is negatively correlated with 'Income' variable. Means increase in one variable leads to decrease on the other one")
```

### **Correlation**

```python
df_corr = df_stat_rev.corr(method ='pearson')
df_stat_rev.corr(method ='pearson')
```

```python
value_corr = df_corr.at["Education_Convert", "Income"]
value_corr
```

```python
if value_corr == 1:
  print("The customer's education is perfectly correlated with the income")
elif 1 > value_corr >= 0.5:
  print("The customer's education is strongly correlated with the income")
elif 0.5 > value_corr >= 0.3:
  print("The customer's education has medium correlation with the income")
elif 0.3 > value_corr > 0:
  print("The customer's education has small correlation with the income")
else:
  print("The customer's education has NO correlation with the income")
```
