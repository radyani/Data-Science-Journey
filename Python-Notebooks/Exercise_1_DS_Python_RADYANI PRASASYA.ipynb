### **Soal 1**: 

Print out integers which are multiples of 3 from 1 to 22.
(Print out bilangan bulat yang merupakan kelipatan 3 dari 1 sampai 22).

**Ekspektasi jawaban:**

Ketika kode kalian di-run, muncul angka-angka: 3, 6, 9, 12, 15, 18, 21. Gunakan looping.

Konsep kode yang digunakan:
1. Angka-angka akan dihasilkan dari fungsi bernama 'kelipatan';
2. Rentang angka yang akan dianalisis dibuat sebagai list terlebih dahulu;
3. Menampilkan list angka yang akan dianalisis;
4. Jika angka di list dibagi dengan angka kelipatan (misalnya 3) tidak memiliki sisa, maka angka tersebut

````python
def kelipatan(angka_min, angka_max, nilai_kelipatan):

    list_angka = list(range(angka_min, angka_max + 1))                            # buat list range angka yang akan dianalisis
    print("List Angka yang Akan Dianalisis:")
    print(list_angka)
    print()

    unit_kelipatan = str(nilai_kelipatan)
    print("Angka Kelipatan " + unit_kelipatan + " di List Tersebut adalah:")

    for angka in list_angka:                                                      # loop mencoba setiap angka dalam list
        if angka % nilai_kelipatan == 0:
          print(angka)
        else:
          pass
````

````python
angka_min = 1           # input batas bawah range
angka_max = 22          # input batas atas range
nilai_kelipatan = 3     # pilih kelipatan berapa

kelipatan(angka_min, angka_max, nilai_kelipatan)
````

### **Soal 2**: 

Buat sebuah function di Python untuk mengecek apakah sebuah angka merupakan bilangan prima atau bukan.

Ekspektasi jawaban: 

misal fungsi Anda bernama prime_check. Ketika kita jalankan prime_check(5), maka outputnya adalah '5 is a Prime number'. Fungsi ini harus bisa berlaku untuk input apa pun. 

Clue: Pengecekan bilangan prima menggunakan looping dan modulo. 
```(python)
print('Hasil dari 5 modulo 2 adalah',5 % 2)  
# 5 Modulo 2 sama dengan 1. Artinya, 5 dibagi 2, sisanya 1

print('Hasil dari 6 modulo 3 adalah', 6 % 3)
# 6 Modulo 3 sama dengan 0. Artinya 6 dibagi 3, sisanya 0.
```

Selamat mengerjakan!

Konsep kode yang digunakan:
1. Fungsi yang digunakan adalah prime_check;
2. Angka yang dimasukkan adalah prima jika tidak bisa dibagi angka selain 1 dan dirinya sendiri tanpa menghasilkan sisa;
3. Angka input akan dibagi dengan pembagi (dimulai dari angka 2 dan seterusnya sampai 1 angka sebelum angka input) 

````python
def prime_check(angka_input):
    
    pembagi = 2
    angka_display = str(angka_input)
    checker = 0 # variabel yang digunakan untuk mengecek berapa kali angka input terdeteksi dapat dibagi

    while pembagi < angka_input:
          if angka_input % pembagi == 0:
              print(angka_display + " is NOT a Prime Number")
              pembagi = angka_input
              checker += 1
          else:
              pembagi += 1

    if checker == 0:
              print(angka_display + " is a Prime Number")  
````

````python
angka_input = 113     # input angka yang akan dicek
prime_check(angka_input)
````
