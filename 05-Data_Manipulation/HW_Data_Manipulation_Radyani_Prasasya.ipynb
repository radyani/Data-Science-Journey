## Part 2

Load Metro_Interstate_Traffic_Volume.csv as DataFrame

From ‘date_time’ column:

*   Create ‘month’ column (10 points)
*   Create ‘day_of_month’ column (digit tanggal) (10 points)
*   Create ‘day_name’ column (Monday, Tuesday, ...) (10 points)
*   Create ‘hour’ column (digit jam) (10 points)
*   Create ‘is_weekend’ binary column (1 if day == saturday OR sunday, 0 otherwise) (10 points) 

```python
df_time = pd.read_csv('Metro_Interstate_Traffic_Volume.csv')
df_time.head(5)
```

```python
df_time['date'] = pd.to_datetime(df_time['date_time'])
df_time = df_time.drop(['date_time'], axis = 1)
df_time.info()
```

```python
#1 Create ‘month’ column
```

```python
df_time['month'] = df_time['date'].dt.month
df_time.head()
```

```python
#2 Create ‘day_of_month’ column (digit tanggal)
```

```python
df_time['day'] = df_time['date'].dt.day
df_time.head()
```

```python
#3 Create ‘day_name’ column (Monday, Tuesday, ...)
```

```python
df_time['day_name'] = df_time['date'].dt.day_name()
df_time.head()
```

```python
#4 Create ‘hour’ column (digit jam) (10 points)
```

```python
df_time['hour'] = df_time['date'].dt.hour
df_time.head()
```

```python
#5 Create ‘is_weekend’ binary column (1 if day == saturday OR sunday, 0 otherwise)
```

```python
df_time["is_weekend"] = df_time['date'].dt.day_name().isin(['Saturday', 'Sunday'])
df_time["is_weekend"] = df_time["is_weekend"].astype(int)
df_time
```
