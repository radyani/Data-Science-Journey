# Assignment

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df.info()
```

# 1. Count of one categorical feature

Bar Chart - Count of Building Types (BldgType)

```python
bldgtype_count = df.groupby(['BldgType'],as_index=False).size()
bldgtype_count = bldgtype_count.sort_values(['size'],ascending=False)
```

```python
bc = sns.barplot(x=bldgtype_count['BldgType'],
            y=bldgtype_count['size'],
            palette = "pastel")
bc.set(xlabel='Building Type', ylabel='Count')
bc.set_title('Frequency of Building Types', fontsize=15, fontweight='bold')
plt.show()
```

Pie Chart - Percentage of Each Building Type Counts per Total Buildings

```python
bt_count = df['BldgType'].value_counts()
btype = bt_count.index
count = bt_count.values
colors = sns.color_palette('pastel')

plt.pie(count, labels=btype,
        colors = colors, 
        autopct='%1.1f%%')
plt.xlabel('Bulding Type')
plt.title("Percentage of Building Type Counts", fontsize=15, fontweight='bold')
plt.show()
```

# 2. Distribution of one continuous feature

Histogram & KDE - Distribution on Lot Area Value (LotArea)

```python
import matplotlib.style as style
sns.set_style("darkgrid")
sns.set(font_scale=1)
histo = sns.displot(x=df['LotArea'],kde=True)
plt.title("Distribution of Lot Area", fontsize=15, fontweight='bold')
plt.xlabel('Lot Area')
```

# 3. Categorical-continuous

Box Plot - Relationship Between House Style (HouseStyle) and Price (SalePrice)



```python
sns.boxplot(x=df['HouseStyle'],
            y=df['SalePrice'])
plt.title("Sale Price According to House Style", fontsize=15, fontweight='bold')
plt.xlabel('House Style', fontsize=13, fontweight='bold')
plt.ylabel('Selling Price', fontsize=13, fontweight='bold')
```

# 4. Continuous-continuous

Scatter Plot - Relationship Between Year the House was Built (YearBuilt) and Price (SalePrice)


```python
sns.set_style("white")
sns.scatterplot(x=df['YearBuilt'],y=df['SalePrice'])
plt.title("Sale Price According to Year House was Built", fontsize=15, fontweight='bold')
plt.xlabel('Year Built', fontsize=13, fontweight='bold')
plt.ylabel('Selling Price', fontsize=13, fontweight='bold')
```
